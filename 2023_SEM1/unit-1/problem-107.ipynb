# Example:

Time series of mean temperature in Ibarra


- The data correspond to time in months versus average (mean) temperature in Ibarra.


- Time period is 3 years.


Carry out interpolation to fill in the gaps in the time series.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.interpolate as interp
```

```python
# Import the data:

data_temp = pd.read_csv("./sample_data/temperature_data.csv")

#print(data_temp)

# now convert to arrays:

month = np.array(data_temp["Month"])
tempe = np.array(data_temp["Mean_temperature(C)"])
```

```python
# Data inspection:

plt.figure()

plt.plot(month, tempe, marker = "+", linestyle = " ")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

```python
# Data inspection:

plt.figure()

plt.plot(month, tempe, linestyle = "-")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

### 1. Using interp1d:

```python
# Interpolaton functions
f_interp1 = interp.interp1d(month, tempe, kind = "linear")
f_interp2 = interp.interp1d(month, tempe, kind = "cubic")

f_interp3 = interp.interp1d(month, tempe, kind = "nearest")
f_interp4 = interp.interp1d(month, tempe, kind = "previous")
f_interp5 = interp.interp1d(month, tempe, kind = "next")
```

```python
# New axis for months:

month_new = np.arange(1, 36.25, 0.25)

#print(month_new)
```

```python
# Evaluate out interpolators

tempe_1 = f_interp1(month_new)
tempe_2 = f_interp2(month_new)

tempe_3 = f_interp3(month_new)
tempe_4 = f_interp4(month_new)
tempe_5 = f_interp5(month_new)
```

```python
# ready to see the results:

# Data inspection:

plt.figure()

plt.plot(month, tempe, marker = "+", linestyle = " ")
plt.plot(month_new, tempe_1, linestyle = "-", color = "green")
plt.plot(month_new, tempe_2, linestyle = "--", color = "magenta")
#plt.plot(month_new, tempe_3, linestyle = "-.", color = "cyan")
#plt.plot(month_new, tempe_4, linestyle = ":", color = "brown")
#plt.plot(month_new, tempe_5, linestyle = "-", color = "green")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

```python
# ready to see the results:

# Data inspection:

plt.figure()

plt.plot(month, tempe, marker = "+", linestyle = " ")
plt.plot(month_new, tempe_3, linestyle = "-.", color = "black")
plt.plot(month_new, tempe_4, linestyle = ":", color = "brown")
plt.plot(month_new, tempe_5, linestyle = "--", color = "green")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

### 2. Using splrep,ev:

```python
# Get the tck object:

tck = interp.splrep(month, tempe, s=0, k=3)

# Evalue the coefficients to get the Splines:

tempe_spline = interp.splev(month_new, tck, der=0)
```

```python
# ready to see the results:

# Data inspection:

plt.figure()

plt.plot(month, tempe, marker = "+", linestyle = " ")
plt.plot(month_new, tempe_spline, linestyle = "-", color = "red")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

### Fill in missing data:

```python
# read_csv returns a DataFrame

data_missing = pd.read_csv("./sample_data/temperature_data_missing.csv")
#print(data_missing )
# To get the data from specific columns, you can use their respective keys, e.g.:

month_missing = np.array(data_missing ["Month"])
tempe_missing = np.array(data_missing ["Mean_temperature(C)"])
```

```python
# Data inspection:

plt.figure()

plt.plot(month_missing, tempe_missing, marker = "+", linestyle = "-.")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

```python
# Interpolaton functions
f_interp_missing = interp.interp1d(month_missing, tempe_missing, kind = "cubic")

# Evaluate 

tempe_missing_1 = f_interp_missing(month_new)

print(tempe_missing_1)
```

```python
# Data inspection:

plt.figure()

plt.plot(month_missing, tempe_missing, marker = "+", linestyle = "-.")
plt.plot(month_new, tempe_missing_1, linestyle = "-")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

### Basic analysis with NANs:

```python
# Mean values:
mean_temp = np.mean(tempe_missing)
std_temp  = np.std(tempe_missing)
print(mean_temp, std_temp)

# Mean values:
mean_temp = np.nanmean(tempe_missing)
std_temp  = np.nanstd(tempe_missing)

print(mean_temp, std_temp)
```

### Remove NANs:

```python
# Get indices for all the finite values

idx_finite = np.isfinite(tempe_missing)

print(idx_finite)
```

```python
# We want to call the not NAN values in both the X axis and Y axis

month_nonan = month_missing[idx_finite] 
tempe_nonan = tempe_missing[idx_finite]

print(tempe_missing)
print(tempe_nonan.shape)

print(month_nonan)
```

### Interpolation using clean data

```python
# Interpolaton functions
f_interp_nonan_linear = interp.interp1d(month_nonan, tempe_nonan\
                                 , kind = "linear")
f_interp_nonan_cubic = interp.interp1d(month_nonan, tempe_nonan\
                                 , kind = "cubic")

# New axis:

month_nonan_new = np.arange(1, 72.25, 0.25)

# Evaluate 
tempe_nonan_linear = f_interp_nonan_linear(month_nonan_new)
tempe_nonan_cubic = f_interp_nonan_cubic(month_nonan_new)

print(tempe_nonan_linear.shape)
```

```python
# Plotting the result:

plt.figure()

plt.plot(month_missing, tempe_missing, marker = "+", linestyle = "-.")
plt.plot(month_nonan_new, tempe_nonan_linear, linestyle = "-")
plt.plot(month_nonan_new, tempe_nonan_cubic, linestyle = "--")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```

### Interpolation using clean data + splines

```python
# Get the tuple
tck_nonan = interp.splrep(month_nonan, tempe_nonan, s=0)

# Call the evaluation function
tempe_nonan_spline = interp.splev(month_nonan_new,\
                                       tck_nonan, der=0)
```

```python
# Plotting the result:

plt.figure()

plt.plot(month_missing, tempe_missing, marker = "+", linestyle = "-.")
plt.plot(month_nonan_new, tempe_nonan_spline, linestyle = "-", color = "brown")

plt.xlabel("Month")
plt.ylabel("Mean Temperature [C]")

plt.show()
```
