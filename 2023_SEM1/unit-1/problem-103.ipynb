## IO python function:

```python
# Start our I/O function

def io_data(filename):
    """
    This is a funtion that reads in a data file
    and returns numpy arrays
    Input: Name of the file we want to open
    Outputs: Data Numpy arrays
    Author: W.E.B.B.
    Date: 09/05/2023
    """
    
    # Create a pandas data frame
    df = pd.read_csv(filename, sep = "\s+")
    
    # Get data into pandas.Series
    time = df["Time"]
    mass = df["Mass1"]
    flxh = df["FluxH"]
    flxb = df["FluxE"]
    
    # Pandas obj to numpy arrays
    time = np.array(time)
    mass = np.array(mass)
    flxh = np.array(flxh) 
    flxb = np.array(flxb)
    
    # Return outputs
    return time, mass, flxh, flxb
    
```

```python
# Start our I/O function

def io_data(filename):
    """
    This is a funtion that reads in a data file
    and returns numpy arrays
    Input: Name of the file we want to open
    Outputs: Data Numpy arrays
    Author: W.E.B.B.
    Date: 09/05/2023
    """
    
    # Create a pandas data frame
    df = pd.read_csv(filename, sep = "\s+")
    
    # Pandas obj to numpy arrays
    time = np.array(df["Time"])
    mass = np.array(df["Mass1"])
    flxh = np.array(df["FluxH"]) 
    flxb = np.array(df["FluxE"])
    
    # Return outputs
    return time, mass, flxh, flxb
    
```

### Basic data analysis

Suppose now that we want to work with the data, not just see them in plots, but actually use them for calculations.

To achieve this Python offers a large range of functions, e.g. say we want to multiply the array of times by 1 minute in seconds, and then save this in a new file jointly with the cloud mass.

For the cloud mass we want to convert g to kg.

```python
# Call our I/O function

t, m, fh, fb = io_data(fn1)
```

```python
# Multiply the time vector by 60 seconds
t_s = t*60. # time t_s in seconds, while t is in mins.

# Multiply the mass vector by 1/1000 kg
m_k = m/1000. # mass m_s is in kg, while m is in grams.
```

# Create pandas data frames:

```python
# New data frame:
# We need the keys and the data:
new_df = pd.DataFrame({"time_s": t_s, "mass_kg": m_k})

print(new_df)
```

```python
# Let's put our data frame into a new file

new_df.to_csv("new_data_1.csv", "," , float_format = "{:.4e}".format)
```

### Open data_2.csv file and output MaxPulse and Calories

```python
# Start our I/O function

def io_data2(filename):
    """
    This is a funtion that reads in a data file
    and returns numpy arrays
    Input: Name of the file we want to open
    Outputs: Data Numpy arrays
    Author: W.E.B.B.
    Date: 09/05/2023
    """
    # Duration,Pulse,Maxpulse,Calories
    # Create a pandas data frame
    df = pd.read_csv(filename, sep = ",")
    
    # Pandas obj to numpy arrays
    du = np.array(df["Duration"])
    pl = np.array(df["Pulse"])
    mp = np.array(df["Maxpulse"]) 
    ca = np.array(df["Calories"])
    
    # Return outputs
    return du, pl, mp, ca
    
```

```python
# Call the function

# Define filename
filename2 = "./sample_data/data_2.csv"

# Same way to define the path: full path
filename2_v2 = "/Users/wladimir/Dropbox/Yachay_Tech/Semestre3_2023/Lectures/computational-physics-1/unit-1/sample_data/data_2.csv"

# Call the function 
dur, pul, mpl, cal = io_data2(filename2_v2)
```
