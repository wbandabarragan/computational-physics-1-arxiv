## Example: Fourier transforms in 1D

(a) Generate 3 sine waves with frequencies $1\rm Hz$, $5 Hz$, and $8 Hz$, amplitudes $4$, $2$ and $0.25$, and phase all zeros.

(b) Plot this $3$ sine waves together with a sampling rate $100 Hz$.

(c) Calculate the DFT of the resulting signal and return the DFT values.

```python
import numpy as np
import matplotlib.pyplot as plt
```

$\chi(k) =\sum_{n=0}^{N-1}x(n)\,e^{\frac{-i\,2\pi\,k\,n}{N}}$

```python
def DFT_function(signal):
    
    N = len(signal)
    n = np.arange(0, N)
    k = n.reshape((N, 1))
    
    exp_term = np.exp(-2j*np.pi*k*n/N)
    
    chi = np.dot(exp_term, signal)
    
    return chi
```

```python
def sine_wave(amp, freq, t):
    # amp  = amplitude
    # freq = frequency
    # t    = time_axis
    
    return amp*np.sin(2*np.pi*freq*t)
```

```python
# sampling rate
s_rate = 100 #Hz

# sampling interval, signal goes from 0 to 1
s_interval = 1.0/s_rate

# t_axis
t = np.arange(0,1, s_interval)

#print(t.size)

# signal amplitudes
amp_1 = 4
amp_2 = 2
amp_3 = 0.25

# signal frequencies
freq_1 = 1
freq_2 = 5
freq_3 = 8

# signal
signal_3 = sine_wave(amp_1, freq_1, t) +\
sine_wave(amp_2, freq_2, t) + sine_wave(amp_3, freq_3, t)

print(t.size, signal_3.size)
```

### Plotting:

```python
plt.figure(figsize= (8,6))

plt.plot(t, signal_3, "red")

plt.xlabel("t(s)")
plt.ylabel("Signal Amplitudes")
plt.show()
```

### Fourier transform - DFT:

```python
fourier_3 = DFT_function(signal_3)

N3 = len(fourier_3)
print(N3)

n3 = np.arange(N3)
print(n3)

T3 = N3/s_rate
print(T3)

freq3 = n3/T3
```

```python
plt.figure(figsize= (8,6))

plt.stem(freq3, np.abs(fourier_3), "b", markerfmt=" ", basefmt="-b")

plt.xlabel(r"$\nu$(Hz)")
plt.ylabel("Fourier Amplitudes")
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.stem(freq3[:N3//2], (abs(fourier_3)[:N3//2])/(N3//2), "b", markerfmt=" ", basefmt="-b")

plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.xlim(0,10)
plt.show()
```
