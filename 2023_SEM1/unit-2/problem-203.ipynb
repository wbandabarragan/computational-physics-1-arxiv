# Finite square well potential:

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
# Fix z_0
z0 = 8

# Vector for z
z = np.arange(0, 10*np.pi, 0.001)
```

$\tan(z) = \sqrt{\left(\frac{z_0}{z}\right)^2 - 1}$

```python
# Evaluate the LHS and RHS

y_lhs = np.tan(z)

y_rhs = np.sqrt((z0**2/z**2) - 1)
```

```python
# Remove large values from the tangent

y_lhs[y_lhs > +4*z0] = np.nan
y_lhs[y_lhs < -4*z0] = np.nan

# Plotting

plt.figure(figsize = (4,3))

plt.plot(z, y_lhs, label = r"$\tan(z)$")
plt.plot(z, y_rhs, label = r"$\sqrt{\left(\frac{z_0}{z}\right)^2 - 1}$")

plt.xlim(0, z0)
plt.ylim(0, 3*z0)

plt.legend()

plt.show()
```

### Solve the equation numerically:


$\frac{\tan(z)}{\sqrt{\left(\frac{z_0}{z}\right)^2 - 1}} - 1= 0$

```python
# Objective function to be optimised

def obj_func(z, z0):
    """
    This is for the root finder.
    """

    opt_func = (np.tan(z))/(np.sqrt((z0**2/z**2) - 1)) - 1

    return opt_func
```

```python
# Optimisation

result = opt.root(obj_func, [1.5, 4, 6], args = z0)
```

```python
print(result)
```

```python
print("z = ", result.x)
```

```python
y1 = np.tan(result.x[0])
y2 = np.tan(result.x[1])
y3 = np.tan(result.x[2])
```

```python
# Remove large values from the tangent

y_lhs[y_lhs > +4*z0] = np.nan
y_lhs[y_lhs < -4*z0] = np.nan

# Plotting

plt.figure(figsize = (6,5))

plt.plot(z, y_lhs)
plt.plot(z, y_rhs)

plt.plot(result.x[0], y1, marker = "o", color = 'red')
plt.plot(result.x[1], y2, marker = "o", color = 'red')
plt.plot(result.x[2], y3, marker = "o", color = 'red')

plt.xlim(0, z0)
plt.ylim(0, 3*z0)

plt.show()
```

### With physical units:

```python
import scipy.constants as cte
```

```python
# Define all constants

hbar = cte.hbar # Planck constant
m = cte.m_e # Mass of e-
a = 2e-10 # half of the width of the well, in m = 2Angstrom = 0.2nm
V0 = 1.60218e-17 # in J = 100eV

# Compute z0
z0 = np.sqrt(2*m*a**2*V0/hbar**2)

print(z0)
```

```python
# Vector for z axis
z = np.arange(0, 4*np.pi, 0.001)

# Functions:
y_lhs = np.tan(z)

y_rhs = np.sqrt((z0/z)**2 - 1)
```

```python
# Plotting

plt.figure(figsize= (6,5))

# We need to cut the tangent
limit_y = 12
y_lhs[y_lhs > limit_y] = np.nan
y_lhs[y_lhs < -limit_y] = np.nan

plt.plot(z, y_lhs, label = "tangent")
plt.plot(z, y_rhs, label = "sqrt")

plt.ylim(0, 12)
plt.legend()
plt.grid()
plt.show()
```

```python
# Optimise, ansatz = educated guess

result = opt.root(obj_func, [1.5, 4, 6.5, 10], args=(z0))
```

```python
# Actual results for z from optimisation

print("Allowed values for z are:\n",  result.x)
```

This potential allows:

Bound states: $E<0$

Scattering states: $E>0$

For bound states, the wavenumbers and energies are quantised:

$tan(z) = \sqrt{\frac{z_0^2}{z^2} -1}$

where $z=la$, $z_0^2= \frac{2ma^2V_0}{\hbar^2}$

$l=\frac{\sqrt{2m(E+V_0)}}{\hbar}$

$k=l\,tan(la) = \frac{\sqrt{-2mE}}{\hbar}$

```python
# Calculate l = z/a, inside the well

print("Allowed values for l in [m^-1] are:\n",  result.x/a)
```

```python
# Calculate k = l*tan(la), outside the well

k_values = ((result.x)/a)*np.tan(result.x)

print("Allowed values for k in [m^-1] are:\n",  k_values)
```

### Allowed Energies:

$E = -\frac{k^2\hbar^2}{2m}$

```python
E_values = -(k_values**2*hbar**2)/(2*m)

print("Allowed values for E in [J] are:\n",  E_values)
```
