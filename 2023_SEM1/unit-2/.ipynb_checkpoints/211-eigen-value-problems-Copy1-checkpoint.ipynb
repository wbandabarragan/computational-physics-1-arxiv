## 2. Finding the largest eigenvalue

Assume $\lambda_1$ is the largest eigenvalue, as $k$ grows the following equation will converge to the first term.

$$ Ax_{k-1} = \lambda_1[v_1+\frac{c_2}{c_1}\frac{\lambda_2^k}{\lambda_1^k}v_2+\dots+\frac{c_n}{c_1}\frac{\lambda_n^k}{\lambda_1^k}v_n] = \lambda_1x_k$$ 

```python
x = np.array([1, 0])
```

```python
A = np.array([[2, 0], [0, 1]])
```

```python
w, v = np.linalg.eig(A)

print(w)
```

```python
iteration = 1

for i in range(iteration):
    x = np.dot(A, x)
    #print(x)
    fac = np.max(abs(x))
    x_n = x / np.max(x)
    print(fac, x_n)
```

## 3. The QR method:

This is a method to find the eigenvalues of a matrix in one go.

We say that 2 square matrices $A$ and $B$ are similar if:

$$A = C^{-1}BC$$

```python
a = np.array([[0, 2], 
              [2, 3]])

q, r = np.linalg.qr(a)
```

```python
print(np.linalg.eig(a))
```

```python
print('Q:\n', q)

print('R:\n', r)
```

```python
b = np.dot(q, r)
print('QR:\n', b)
```

```python
p = [1, 5, 10, 20]

for i in range(20):
    q, r = np.linalg.qr(a)
    a = np.dot(r, q)
    if i+1 in p:
        print(f'Iteration {i+1}:')
        print(a)
```

### Problem: QM

```python
h_matrix = np.array([[2, 1, 0], 
              [1, 2, 0],
              [0, 0, 3]])
w,v=np.linalg.eig(h_matrix)
print('E-value:\n', w)
print('E-vector:\n', v)
```

```python
psi = np.array([1.j/np.sqrt(3), -1.j/np.sqrt(3), 1.j/np.sqrt(3)])

print(psi)
```

```python
p1 = np.dot(v[0],psi)**2/np.dot(psi, psi)
p2 = np.dot(v[1],psi)**2/np.dot(psi, psi)
p3 = np.dot(v[2],psi)**2/np.dot(psi, psi)
print(p1, p2, p3)
```

```python
H_exp = np.dot(np.dot(np.conjugate(psi),h_matrix), psi.T)

print(H_exp)
```

```python
H_exp_2 = np.dot(np.dot(np.conjugate(psi),h_matrix@h_matrix), psi.T)

print(H_exp)
```

```python
sigma_H = np.sqrt(H_exp_2 - H_exp**2)

print(sigma_H)
```
