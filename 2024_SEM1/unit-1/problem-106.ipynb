# Exercise: Fitting spectral lines

In this exercise we will be fitting Gaussian functions to spectral lines of the $\rm H_I$ transition of the Hydrogen atom. This is a forbidden transition that occurs as a result of its hyperfine structure.


Let us assume the spin of the p+ is down, then the $\rm H_I$ transition occurs when the spin of the electron flips from up to down (i.e. when going from anti-parallel to parallel). The spin up configuration has a slightly higher energy than the spin down because the magnetic dipole moments are parallel, so when it flips it releases a low-energy photon and goes into the spin down configuration, which has lower energy as the the magnetic dipole moments are anti-parallel.


**And some nice extra details from Wikipedia:**

"This transition is highly forbidden with an extremely small transition rate of $2.9\times 10^{-15}\,\rm s^{-1}$ and a mean lifetime of the excited state of around 10 million years. A spontaneous occurrence of the transition is unlikely to be seen in a laboratory on Earth, but it can be artificially induced using a hydrogen maser. It is commonly observed in astronomical settings such as hydrogen clouds in our galaxy and others. Because of its long lifetime, the line has an extremely small natural width, so most broadening is due to Doppler shifts caused by bulk motion or nonzero temperature of the emitting regions."

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as opt
```

### 1. Data I/O:

```python
# Let's place the data into a dataframe
HI_spectrum = pd.read_csv("../sample-data/104444.9-264831_gass_spectrum.txt", sep = "\s+")

#print(HI_spectrum)
```

```python
# Pass dataframe to arrays
velocity  = np.array(HI_spectrum["velocity"])
intensity = np.array(HI_spectrum["Tb"])

# Check
#print(intensity)
```

#### 2. Spectrum plot:

```python
# Plotting

plt.figure()

plt.plot(velocity, intensity, label = r"$21cm$ spectrum")

plt.xlabel(r"$v$ [km/s]")
plt.ylabel(r"$T_b$")

plt.show()
```

```python
# Plotting

plt.figure()

plt.plot(velocity, intensity, label = r"$21cm$ spectrum")

plt.xlabel(r"$v$ [km/s]")
plt.ylabel(r"$T_b$")

plt.xlim(-50, +50)

plt.show()
```

#### 3. We propose multi-Gaussian model:

```python
# Python function to model the Gaussian

def gaussian_function(x, a, b, c):
    """
    This defines the Gaussian model for the regression.
    Inputs: x -> velocity vector, a is the height,
            b is the width, c is the displacement 
    Output: y -> Gaussian f.
    """
    
    y = a*np.exp(-b*(x + c)**2)
    
    return y
```

#### 4. Optimisation

We want to find the $a, b, c$ parameters that best firt the data.

We use curve_fit().

```python
# The regression coefficients and the covariance matrix.

coef, cova = opt.curve_fit(gaussian_function, velocity, intensity)

 is

#print(cova)

print("Associated uncertainties:", np.sqrt(np.diag(cova)))
```

#### 4. Overplot the fit on the original spectrum

```python
# Call the Gaussian function

intensity_fit = gaussian_function(velocity, *coef)
```

```python
# Plotting

plt.figure()

plt.plot(velocity, intensity, label = r"$21cm$ spectrum")

plt.plot(velocity, intensity_fit, label = r"Best fit")

plt.xlabel(r"$v$ [km/s]")
plt.ylabel(r"$T_b$")

plt.xlim(-50, +50)

plt.show()
```

#### 5. New regression model:

Clearly the previous 1-Gaussian model was not good enough, we need 2 Gaussians.



```python
# Python function to model the Gaussian

def two_gaussian_function(x, a, b, c, d, e, f):
    """
    This defines the Gaussian model for the regression.
    Inputs: x -> velocity vector, a is the height,
            b is the width, c is the displacement 
    Output: y -> Gaussian f.
    """
    
    y = a*np.exp(-b*(x + c)**2) + d*np.exp(-e*(x + f)**2)
    
    return y
```

#### 6. New optimisation:

```python
# The regression coefficients and the covariance matrix.

coef2, cova2 = opt.curve_fit(two_gaussian_function, velocity, intensity)

print("Best-fit parameters", coef2)

#print(cova)

print("Associated uncertainties:", np.sqrt(np.diag(cova2)))
```

#### 7. Overplot our new fit on the original spectrum

```python
# Call the Gaussian function

intensity_fit2 = two_gaussian_function(velocity, *coef2)

# Get each Gaussian separately
gaussian_1 = gaussian_function(velocity, coef2[0], coef2[1], coef2[2])
gaussian_2 = gaussian_function(velocity, coef2[3], coef2[4], coef2[5])
```

```python
# Now we plot

# Plotting

plt.figure()

plt.plot(velocity, intensity, label = r"$21cm$ spectrum")

plt.plot(velocity, intensity_fit2, label = r"Best fit", color = "magenta", linestyle = "-")

# Plot each Gaussian
plt.plot(velocity, gaussian_1, label = r"Gaussian 1", color = "green", linestyle = ":")
plt.plot(velocity, gaussian_2, label = r"Gaussian 2", color = "red", linestyle = "-.")

plt.xlabel(r"$v$ [km/s]")
plt.ylabel(r"$T_b$")

plt.xlim(-50, +50)

plt.show()
```

#### 8. Relative velocity between the clouds:

We have to compare the displacement parameters (c, f):

```python
print(" The relative velocity (in km/s) between the two clouds is:", abs(coef2[2] - coef2[5]))
```

# 2nd way: Fitting using astropy

See documentation: https://docs.astropy.org/en/stable/api/astropy.modeling.fitting.LevMarLSQFitter.html

```python
import astropy.modeling as mod
```

```python
# We need to define our fitter, which uses the LM method
fitter = mod.fitting.LevMarLSQFitter()

# Next we need the model
model = mod.models.Gaussian1D()

# Call the fitted model
fitted_model = fitter(model, velocity, intensity)

print(fitted_model)

# How you call the best fit data
#print(fitted_model(velocity))
```

```python
# Plotting

plt.figure()

plt.plot(velocity, intensity, label = r"$21cm$ spectrum")

plt.plot(velocity, fitted_model(velocity), label = r"Best fit")

plt.xlabel(r"$v$ [km/s]")
plt.ylabel(r"$T_b$")

plt.xlim(-50, +50)

plt.show()
```
