### Problem 1 (Chemical reactions):

The following chemical reaction is unbalanced:

$C_2H_6 + O_2\to CO_2 + H_2O$

Our job is to find $(洧논_1,洧논_2,洧논_3,洧논_4)$, so that the above equation is balanced:

$x_1\,C_2H_6 + x_2\,O_2\to x_3\,CO_2 + x_4\,H_2O$

```python
import sympy as sp
sp.init_printing()
import numpy as np 
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scienceplots as splt
```

#### (a) Write down the system of linear equations, so that each element is balanced:

- Carbon (C)

- Hydrogen (H)

- Oxygen (O)

Carbon: $2 x_1 - x_3 = 0$

Hydrogen: $6x_1 - 2x_4 = 0$

Oxygen: $2x_2 - 2x_3 -x_4 = 0$

#### (b) Create a python routine that finds the parametric solution to the system of linear equations.



```python
# Define symbols 
x = sp.symbols('x1:5', subs = True)

# Define the equations 
C_eq = sp.Eq(2 * x[0], x[2])
H_eq = sp.Eq(6 * x[0], 2 * x[3])
O_eq = sp.Eq(2 * x[1], 2 * x[2] + x[3])

# Let's define a function to solve this system algebraically.
param_sol = lambda eq1, eq2, eq3, x: sp.solve((eq1, eq2, eq3), x)
```

```python
# Call the function 
solution = param_sol(C_eq, H_eq, O_eq, x)

# Print the solution
print(solution)
```

#### (b) Create a Python function to generate circumferences centred in the origin of a 2D Cartesian plane. The function should return the $(x,y)$ coordinates of circumferences given a value for the radius and a vector containing values for the angle $\theta \in [0, 2\,\pi]$. The latter can be defined as a global parameter. Recall that the coordinates of a circumference can be obtained as follows:

$$x = r\,\cos(\theta)$$

$$y = r\,\sin(\theta)$$

```python
def circ_f(radius, angle):
    """
    """
    # define the coordinates of circumferences 
    x_values = radius * np.cos(angle)
    y_values = radius * np.sin(angle) 
    
    return x_values, y_values 
```
