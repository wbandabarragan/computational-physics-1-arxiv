# Simulation analysis:

- Choose one of the files (e.g. #10) and make a plot of density + velocity vector field.


- Loop over the files and make the same plot (density + velocity) for all the files in the simulation.


- Plot: density + velocity vector fields


- Simulation fields: Convert the quantities in code units (rho, vx1, vx2, Bx1, Bx2) to CGS units, for which we need to read the file: **units.out** which is placed in the TUR-DRIVE-SUP folder.


To do this, you can multiply the variables by the normalisation parameter in CGS units, e.g. to get the density in CGS units:


$\rho_{cgs} = \rho\times\rho_0$, where $\rho$ comes from the simulation and $\rho_0$ comes from the **units.out** file.


$L_{cgs} = L\times L_0$, where $L$ is the length of the box in code units and $L_0$ comes from the **units.out** file.


To get the times in CGS units, we need to read the second column of **vtk.out**.


**Notes:**


- All the plots should be produced in CGS units.


- An animation in **GIF** should be produced at the end.

## 1. Import libraries 

```python
import pyvista as pv

import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

import os # this is to create a folder for the figures.
```

## 2. Read mesh:

```python
mesh = pv.read("./TURB_DRIVE_SUP/data.0010.vtk")
```

```python
# In case you want to plot interactively

#mesh.plot()
```

## 3. Get data fields from arrays:

```python
rho = pv.get_array(mesh, "rho", preference = 'cell') #density in code units
vx1 = pv.get_array(mesh, "vx1", preference = 'cell') #vx in code units
vx2 = pv.get_array(mesh, "vx2", preference = 'cell') #vy in code units
Bx1 = pv.get_array(mesh, "Bx1", preference = 'cell') #Bx in code units
Bx2 = pv.get_array(mesh, "Bx2", preference = 'cell') #By in code units
```

```python
print(rho.shape)
```

## 4. Reshape 1D arrays into 2D arrays

```python
rho_2D = rho.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
vx1_2D = vx1.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
vx2_2D = vx2.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
Bx1_2D = Bx1.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
Bx2_2D = Bx2.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
```

```python
print(Bx1_2D.shape)
```

## 5. Unit normalisation: Code units to CGS units

### Open the units.out file (CSV file)

```python
units_data = pd.read_csv("./TURB_DRIVE_SUP/units.out")
```

```python
print(units_data)
```

### Test code to check where the values are:

```python
A = units_data.loc[units_data["variable"] == "rho_0"]["normalisation"]
print(units_data.loc[units_data["variable"] == "rho_0"]["normalisation"])

print(type(A))

B = np.array(A)

print(B)
```

### Clean code to obtain normalisation values:

```python
rho_00 = np.array(units_data.loc[units_data["variable"] == "rho_0"]["normalisation"])
vel_00 = np.array(units_data.loc[units_data["variable"] == "v_0"]["normalisation"])
len_00 = np.array(units_data.loc[units_data["variable"] == "L_0"]["normalisation"])

# The last one is the isothermal sound speed
c_iso00 = np.array(units_data.loc[units_data["variable"] == "c_iso"]["normalisation"])
```

```python
print(type(rho_00))
print(vel_00)
print(len_00)
print(c_iso00)
```

### Normalising variables to CGS units:

```python
rho_cgs2D = rho_2D*rho_00
vx1_cgs2D = vx1_2D*vel_00
vx2_cgs2D = vx2_2D*vel_00
```

## 6. Plotting:

### Creating a mesh with the correct sizes in CGS units:

```python
xx = np.linspace(mesh.bounds[0]*len_00, mesh.bounds[1]*len_00, mesh.dimensions[0] - 1) # in cm

yy = np.linspace(mesh.bounds[2]*len_00, mesh.bounds[3]*len_00, mesh.dimensions[0] - 1) # in cm
```

```python
XX, YY = np.meshgrid(xx, yy)
```

### Plotting density field in CGS units

```python
plt.figure(figsize = (12,10))

Z = plt.contourf(XX, YY, rho_cgs2D, 30)

plt.colorbar(Z)

plt.show()
```

```python
plt.figure(figsize = (12,10))

Z = plt.pcolor(XX, YY, rho_cgs2D, shading='auto')

plt.colorbar(Z)

plt.show()
```

### Plotting the velocity map, for which we calculate the modulus in CGS units:

```python
modulus_velocity = np.sqrt(vx1_cgs2D**2 + vx2_cgs2D**2)
```

### And the plot using quiver:

```python
plt.figure(figsize = (12,10))

Z = plt.quiver(XX, YY, vx1_cgs2D, vx2_cgs2D, modulus_velocity, cmap = 'magma')

plt.colorbar(Z)

plt.show()
```

## Combined plot with customisations:

```python
plt.figure(figsize = (16,10))

Z1 = plt.pcolor(XX, YY, rho_cgs2D, cmap = 'viridis', shading = 'auto', vmin = 0, vmax= 4.e-24)
Z2 = plt.quiver(XX, YY, vx1_cgs2D, vx2_cgs2D, modulus_velocity, cmap = 'Oranges', clim = (0,3e6))

CB1 = plt.colorbar(Z1) # Create colour bar object 1
CB2 = plt.colorbar(Z2) # Create colour bar object 2

# Add labels to colour bar
CB1.set_label('Density ($g/cm^3$)', size=18) 
CB2.set_label('Velocity ($cm/s$)', size=18)

# Change font size of ticks in colour bar
CB1.ax.tick_params(labelsize=18)
CB2.ax.tick_params(labelsize=18)

# Change font size of e^() values on top of the colourbars
CB1.ax.yaxis.get_offset_text().set(size=18)
CB2.ax.yaxis.get_offset_text().set(size=18)

# Change font size of ticks in the grid map
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

# Change font size of the normalisation values
plt.rc('font', size=18)

# Display plot
plt.show()
```

## 7. Loop over files:

```python
import os
```

```python
for i in range(0,101):
    
    #print(i)
    
    filename = "./TURB_DRIVE_SUP/data.0{:03d}.vtk".format(i)
    
    #print(filename)
    
    mesh = pv.read(filename)
    
    rho = pv.get_array(mesh, "rho", preference = 'cell') #density in code units
    vx1 = pv.get_array(mesh, "vx1", preference = 'cell') #vx in code units
    vx2 = pv.get_array(mesh, "vx2", preference = 'cell') #vy in code units
   
    rho_2D = rho.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    vx1_2D = vx1.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    vx2_2D = vx2.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    
    rho_cgs2D = rho_2D*rho_00
    vx1_cgs2D = vx1_2D*vel_00
    vx2_cgs2D = vx2_2D*vel_00

    if os.path.isdir("./TURB_DRIVE_SUP/Figures"):
        print("Directory already exists.")
    else:
        print("Directory is being created.")
        os.mkdir("./TURB_DRIVE_SUP/Figures")
    
    xx = np.linspace(mesh.bounds[0]*len_00, mesh.bounds[1]*len_00, mesh.dimensions[0] - 1) # in cm
    yy = np.linspace(mesh.bounds[2]*len_00, mesh.bounds[3]*len_00, mesh.dimensions[0] - 1) # in cm
    
    XX, YY = np.meshgrid(xx, yy)
    
    plt.figure(figsize = (16,10))

    Z1 = plt.pcolor(XX, YY, rho_cgs2D, cmap = 'viridis', shading = 'auto', vmin = 0, vmax= 4.e-24)
    Z2 = plt.quiver(XX, YY, vx1_cgs2D, vx2_cgs2D, modulus_velocity, cmap = 'Oranges', clim = (0,3e6))

    CB1 = plt.colorbar(Z1)
    CB2 = plt.colorbar(Z2)

    CB1.set_label('Density ($g/cm^3$)', size=18)
    CB2.set_label('Velocity ($cm/s$)', size=18)

    CB1.ax.tick_params(labelsize=18)
    CB2.ax.tick_params(labelsize=18)

    CB1.ax.yaxis.get_offset_text().set(size=18)
    CB2.ax.yaxis.get_offset_text().set(size=18)

    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)

    plt.rc('font', size=18)

    plt.savefig("./TURB_DRIVE_SUP/Figures/rho-vel0{:03d}.png".format(i))
    plt.close()
      
```

```python
import glob
from PIL import Image
```

```python
images_in = "./TURB_DRIVE_SUP/Figures/rho-vel****.png"

gif_image_out = "./TURB_DRIVE_SUP/Figures/animation-rho-vel.gif"
```

```python
imgs = (Image.open(f) for f in sorted(glob.glob(images_in)))

img = next(imgs)

img.save(fp = gif_image_out, format='GIF', append_images=imgs, save_all=True, duration=100, loop=0)
```
