## Particle in a finite potential well

We need to find the energy of the bound states levels given the following transcendental equation:

$\tan(z) = \sqrt{\frac{z^2_0}{z^2} -1}$ 


where:

$z^2_0 = \frac{2\,m\,a^2\,V_0}{\hbar^2}$


$z = la = \frac{\sqrt{2m(E+V_0)}}{\hbar}\,a$

```python
import numpy as np

import matplotlib.pyplot as plt

import scipy.optimize as opt
```

```python
z = np.arange(0, 10*np.pi, 0.0001)
```

```python
z_0 = 8
```

```python
y1 = np.tan(z)

y2 = np.sqrt((z_0**2/z**2) - 1)
```

```python
y1[y1 > +3*z_0] = np.nan
y1[y1 < -3*z_0] = np.nan
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(z, y1)
ax.plot(z, y2)

plt.xlim(0, 10)
plt.ylim(0, 20)
plt.show()
```

## Solution using root

```python
def tan_function(z, z_0):
    return (np.tan(z)/np.sqrt((z_0**2/z**2) - 1)) - 1
```

```python
result = opt.root(tan_function, [1.3, 4.0, 7.0], args = z_0)
```

```python
print(result)
```

```python
print(result.x)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(z, y1)
ax.plot(z, y2)
ax.plot(result.x[0], np.tan(result.x[0]), marker = 'd')
ax.plot(result.x[1], np.tan(result.x[1]), marker = 'd')
ax.plot(result.x[2], np.tan(result.x[2]), marker = 'd')

plt.xlim(0, 10)
plt.ylim(0, 20)
plt.show()
```
