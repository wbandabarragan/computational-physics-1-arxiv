### Basic plotting with pyplot

- Pyplot is an interface to plot using Matlab- and Mathematica-like plotting sequences.
- It facilitates plotting.

Plotting can be done without using "objects", e.g.

```python
# Line added to see plots inline
%matplotlib inline

# Importe libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate two vectors x and y(x), where y(x) is a function.
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

Or we can use objects, e.g.

```python
# Example using plt.figure, which returns one tuple with one object

fig = plt.figure()
ax = fig.add_subplot(111) # 1 row, 1 column, 1 location
ax.plot(x, y)
```

In the example above, plt.figure() is a function that returns a "tuple".

This tuple contains one "object":

- Object 1: figure


We can also write the same in one line, using plt.subplots()

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

In the example above, plt.subplots() is a function that returns a "tuple".

This tuple contains two "objects":

- Object 1: figure
- Object 2: axes

## Example (plotting functions):

Let us plot a vector against itself.

For this ax.plot() should be invoked.

```python
#generate some data to plot:
x = np.arange(0,10,0.1)

# plot the data against itself:
fig, ax = plt.subplots()
ax.plot(x, x)
plt.show()
```

And a vector against a cuadratic function of itself:

```python
# plot the data against the ^2 of itself:
fig, ax = plt.subplots()
ax.plot(x, x**2, c='red', lw=5.0)
plt.show()
```

## Example (Overplotting):

Let's now plot two functions, $y(x)$ and $z(x)$, on the same space:

```python
x = np.arange(-3*np.pi,3*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y, c='green')
ax.plot(x, z, c='brown')
plt.show()
```

## Example (Customisation):


- You can customise the size of the figure with figsize=( , ). 

- Add subplots with fig.add_subplot(# rows,# columns,plot order).

- Specify colours with c = plues the standard C_ notation: https://matplotlib.org/2.0.2/users/colors.html

- Modify the linestyle with linestyle=' '

- Add a general title with: plt.title()

- Add Latex with an r'' enviroment in titles or axes.

- Add labels in ax.plot().

- Print the labels with ax.legend().

```python
y2 = np.sin(x+np.pi/2)
z2 = np.cos(x+np.pi/2)

fig = plt.figure(figsize=(10,4))  #custom figure size
ax = fig.add_subplot(1,2,1)  # to create multiple subplots
ax.plot(x, y, c='C0', linestyle='-', label=r'$\sin(x)$')
ax.plot(x, y2, c='C1', linestyle=':', label=r'$\sin(x+\frac{\pi}{2})$')
ax.legend(loc=1)
plt.title(r'$\sin(x)$ and $\sin(x + \frac{\pi}{2})$')

ax = fig.add_subplot(1,2,2)
ax.plot(x, z, c='C2', linestyle = '--', label=r'$\cos(x)$')
ax.plot(x, z2, c='C3', linestyle='-.', label=r'$\cos(x+\frac{\pi}{2})$')
ax.legend(loc=4)
plt.title(r'$\cos(x)$ and $\cos(x + \frac{\pi}{2})$')
plt.show()
```

## Example (scatter plots):

- This is useful when you have two vectors x and y containing data.

```python
# Create some random numbers
np.random.seed(12) # Fixing random state for reproducibility

x = np.random.randn(100)
y = np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(x, y, marker='d', linestyle='', c='brown') 

plt.show()
```
